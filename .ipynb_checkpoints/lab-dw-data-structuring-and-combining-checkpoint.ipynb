In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

```python
# Your code goes here
import pandas as pd
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
df1
```

```python
df2.head()
```

```python
df3.head()
```

```python
df1.columns = [col.lower().replace(" ","_") for col in df1.columns]
df1.rename(columns= {'st': 'state'}, inplace=True)
df1.columns
```

```python
df2.columns = [col.lower().replace(" ","_") for col in df2.columns]
df2.rename(columns= {'st': 'state'}, inplace=True)
df2.columns
```

```python
df3.columns = [col.lower().replace(" ","_") for col in df3.columns]
df3.columns
```

```python
df = pd.concat([df1, df2, df3], axis=0, join="inner")
df.columns
```

```python
display(df)
```

```python
df.nunique()
```

```python
df.customer.unique()
```

```python
df.customer.nunique()
```

```python
df.duplicated(subset=['customer']).sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.gender.unique()
```

```python
df['gender'] = df['gender'].str.strip().str.lower().replace({
    'male': 'M', 
    'female': 'F', 
    'femal': 'F',
    'f':'F',
    'm':'M'
})
```

```python
df.gender.unique()
```

```python
df['state'].unique()
```

```python
df['state'] = df['state'].replace({
    'AZ': 'Arizona', 
    'WA': 'Washington',
    'Cali': 'California'
})
```

```python
df['state'].unique()
```

```python
df['education'].unique()
```

```python
df['education'] = df['education'].replace({
    'Bachelors': 'Bachelor'
})
```

```python
df['education'].unique()
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace("%","")
```

```python
df['customer_lifetime_value'].unique()
```

```python
df['vehicle_class'].unique()
```

```python
df['vehicle_class'] = df['vehicle_class'].replace({
    'Luxury SUV': 'Luxury', 
    'Sports Car': 'Luxury',
    'Luxury Car': 'Luxury'
})
```

```python
df['vehicle_class'].unique()
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(float)
```

```python
df['number_of_open_complaints'].unique()
```

```python
def transform_complaints(value):
    if isinstance(value, str):
        return int(value.split('/')[1])
    return value
```

```python
df['number_of_open_complaints'] = df['number_of_open_complaints'].apply(transform_complaints)
```

```python
df['number_of_open_complaints'].unique()
```

```python
df_numerical = df.select_dtypes(include='number')
df_categorical = df.select_dtypes(include='object')
```

```python
df_categorical
```

```python
df_categorical.dropna()
```

```python
avg_customer_lifetime_value = df.customer_lifetime_value.mean()
avg_income = df.income.mean()
avg_monthly_premium_auto = df.monthly_premium_auto.mean()
avg_number_of_open_complaints = df.number_of_open_complaints.mean()
avg_total_claim_amount = df.total_claim_amount.mean()
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].fillna(avg_customer_lifetime_value)
df["income"] = df["income"].fillna(avg_income)
df["monthly_premium_auto"] = df["monthly_premium_auto"].fillna(avg_monthly_premium_auto)
df["number_of_open_complaints"] = df["number_of_open_complaints"].fillna(avg_number_of_open_complaints)
df["total_claim_amount"] = df["total_claim_amount"].fillna(avg_total_claim_amount)
```

```python
df.isna()
```

```python
display(df)
```
